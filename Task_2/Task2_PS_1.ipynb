### Linear Regression for CTC Predictions

Importing the necessary library

```python
import pandas as pd
import numpy as np
import seaborn as sns
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error as mscore
```

Importing the data as Dataframe

```python
df = pd.read_csv('Data_file.csv')
```

```python
df_1 = pd.read_csv('Test_data_file.csv')
```

```python
df.head(10)
```

Checking for null value

```python
df.isnull().any()
```

```python
df.drop('S.No.',axis=1,inplace=True)
```

Converting Catagorical values to Numberical values

```python
df = pd.get_dummies(df,columns=['College','Role','City type'], drop_first=True)
```

```python
df
```

As model can't convert to float, converting the datatype of CTC and Previous CTC to float

```python
df['CTC'] = df['CTC'].str.replace(',','')
```

```python
df['Previous CTC'] = df['Previous CTC'].str.replace(',','')
```

```python
df['CTC']=df['CTC'].astype('float')
df['Previous CTC']=df['Previous CTC'].astype('float')
```

Separting Features and Labels

```python
y = df.iloc[:,4:5]
```

```python
X = df.drop(columns=['CTC'])
```

```python
model = LinearRegression()
```

Fitting Model

```python
model.fit(X,y)
```
